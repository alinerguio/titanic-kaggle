## Titanic - Machine Learning from Disaster

[Desafio](https://www.kaggle.com/c/titanic/data) do Kaggle para descobrir se um passageiro sobreviveu ao desastre a partir de suas características.

Utilizamos as bibliotecas de análise e estruturas de dado [pandas](https://pandas.pydata.org/) e [numpy](https://numpy.org/), ferramenta de predição [sklearn](https://scikit-learn.org/stable/) - com o algoritmo [KNN](https://en.wikipedia.org/wiki/K-nearest_neighbors_algorithm). Além disso, a biblioteca [seaborn](https://seaborn.pydata.org/) para alguns plots básicos para avaliação do resultado.

### Dados disponíveis



Fonte: https://www.kaggle.com/c/titanic/data

```python
import pandas as pd
import numpy as np
from sklearn.neighbors import KNeighborsClassifier
```

```python
treinamento = pd.read_csv('data/train.csv')
teste = pd.read_csv('data/test.csv')
```

```python
treinamento.head()
```

```python
len(treinamento)
```

```python
len(treinamento.Ticket.unique())
```

```python
len(treinamento.Cabin.unique())
```

```python
len(teste.Cabin.unique())
```

```python
treinamento_set = set(treinamento.Cabin.unique())
```

```python
teste_set = set(teste.Cabin.unique())
```

```python
len(treinamento_set & teste_set)
```

```python
treinamento_sex = pd.get_dummies(treinamento.Sex)
treinamento_sex.head()
```

```python
treinamento_embarked = pd.get_dummies(treinamento.Embarked)
treinamento_embarked.head()
```

```python
X = treinamento.drop(columns=['PassengerId','Survived', 'Name', 'Ticket', 'Cabin'])
```

```python
X.head()
```

```python
X = pd.concat([X, treinamento_sex, treinamento_embarked], axis=1).drop(columns=['Sex', 'Embarked'])
```

```python
X = X.replace(np.nan, 0)
```

```python
X.head()
```

```python
X = X.to_numpy()
```

```python
X
```

```python
y = treinamento[['Survived']].to_numpy()
```

```python
neigh = KNeighborsClassifier(n_neighbors=7)
neigh.fit(X, y)
```

```python
teste.head()
```

```python
para_testar = teste.drop(columns=['PassengerId','Name', 'Ticket', 'Cabin'])
teste_sex = pd.get_dummies(para_testar.Sex)
teste_sex.head()
teste_embarked = pd.get_dummies(para_testar.Embarked)
teste_embarked.head()
para_testar = pd.concat([para_testar, teste_sex, teste_embarked], axis=1).drop(columns=['Sex', 'Embarked'])
```

```python
para_testar.head()
```

```python
para_testar = para_testar.replace(np.nan, 0.0)
```

```python
para_testar = para_testar.to_numpy()
```

```python
resultado = neigh.predict(para_testar)
```

```python
resultado
```

```python
teste.PassengerId.to_numpy()
```

```python
enviar = pd.DataFrame(data={'PassengerId':teste.PassengerId.to_numpy(), 'Survived': resultado})
```

```python
len(enviar)
```

```python
enviar.head()
```

```python
enviar.to_csv('resultado_titanic.csv', index=False)
```
